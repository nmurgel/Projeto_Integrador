```python
import pandas as pd
import requests as req

# DataSet - Ibama Dados Abertos

res = req.get('http://siscom.ibama.gov.br/geoserver/publica/ows?service=WFS&version=1.0.0&request=GetFeature&typeName=publica:veg_naofloresta_aml_a&outputFormat=application%2Fjson')

res.json()
```

```python
val = res.json()['features']
df = pd.DataFrame(val)
df.columns
```
